# Mean encoding


```python


def _getMeanEncodingColumnName(groupby, target, agg):
    if len(groupby) == 1:
        column_name = groupby[0]
    else:
        column_name = '_'.join(groupby)

    return column_name + '_' + agg + '_' + target

def _getMeanEncodingCorrCoef(data, new_feature, target):
    t = data[~data[target].isnull()]
    encoded_feature = t[new_feature].values
    corr = np.corrcoef(t[target].values, encoded_feature)[0][1]
    return corr

def _getMeanEncodingBaseValue(data, target, agg):
    base = np.nan
    if agg == 'mean':
        base = data[target].mean()
    elif agg == 'sum':
        base = 0
    elif agg == 'std':
        base = data[target].std()
    elif agg == 'var':
        base = data[target].var()
    
    return base

def _ApplyAggregate(data, val_data, groupby, target, column_name, agg):
    base = _getMeanEncodingBaseValue(data, target, agg)
    if column_name == None:
        column_name = _getMeanEncodingColumnName(groupby, target, agg)
    data[column_name] = data.groupby(groupby)[target].transform(agg)
    #Fill NaNs with base value
    data[column_name].fillna(base, inplace=True)
    #Calculate correlation with target variable
    score = _getMeanEncodingCorrCoef(data, column_name, target)
    val_data = val_data.merge(data.groupby(groupby, as_index=False).agg({column_name : 'first'}), on=groupby, how='left')
    return data, val_data, column_name, score

# data: training dataset
# val_data: validation dataset
# groupby: list of columns to group by for mean encodings
# target: the name of the target variable
# column_name (optional): the name of the new feature (it will be infered if None)
# it will return a tuple (train_dataset, val_dataset, column_name, correlation_score)
def MeanEncoding(data, val_data, groupby, target, column_name=None):
    agg = 'mean'
    return _ApplyAggregate(data, val_data, groupby, target, column_name, agg)

def SumEncoding(data, val_data, groupby, target, column_name=None):
    agg = 'sum'
    return _ApplyAggregate(data, val_data, groupby, target, column_name, agg)

def StdEncoding(data, val_data, groupby, target, column_name=None):
    agg = 'std'
    return _ApplyAggregate(data, val_data, groupby, target, column_name, agg)

def VarEncoding(data, val_data, groupby, target, column_name=None):
    agg = 'var'
    return _ApplyAggregate(data, val_data, groupby, target, column_name, agg)


def MeanEncodingKFold(data, val_data, groupby, target, splits=5, shuffle=False, column_name=None, debug=True):
    
    if (debug):
        print("[DEBUG] Creating mean encoding for " + target + " grouped by " + "_".join(groupby) + " using " + str(splits) + " folds")

    prior = data[target].mean() #global mean
    if column_name == None:
        column_name = _getMeanEncodingColumnName(groupby, target, 'mean')
    
    # initialize empty column
    original_column_list = data.columns
    data[column_name] = np.nan
    kf = KFold(n_splits=splits, shuffle=shuffle)
    splits = kf.split(data)

    if (debug):
        cf = 1
    for tr_ind, val_ind in splits:
        if (debug):
            print("[DEBUG] Processing Fold ", cf)

        X_tr, X_val = data.iloc[tr_ind][original_column_list], data.iloc[val_ind][original_column_list]
        g = X_tr.groupby(groupby,as_index=False).agg({target:'mean'}).rename(columns={target : column_name})
        original_index = X_val.index
        X_val = pd.merge(X_val, g, on=groupby,how='left')
        X_val.index = original_index
        data.iloc[val_ind] = X_val

        if (debug):
            cf = cf+1
     
    #Fill NaNs with global mean
    data[column_name].fillna(prior, inplace=True)
    #Calculate relevance
    score = _getMeanEncodingCorrCoef(data, column_name, target)
    val_data = val_data.merge(data.groupby(groupby, as_index=False).agg({column_name : 'mean'}), on=groupby, how='left')
    return data, val_data, column_name, score

def MeanEncodingLOO(data, val_data, groupby, target, column_name=None):
    prior = data[target].mean() #global mean
    if column_name == None:
        column_name = _getMeanEncodingColumnName(groupby, target, 'mean')
    data[column_name] = data.groupby(groupby)[target].transform(lambda x: (x.sum() - x) / (x.count() - 1))
    # Fill NaNs
    data[column_name].fillna(prior, inplace=True) #fill NaNs with global mean
    score = _getMeanEncodingCorrCoef(data, column_name, target)
    val_data = val_data.merge(data.groupby(groupby, as_index=False).agg({column_name : 'mean'}), on=groupby, how='left')
    return data, val_data, column_name, score

def MeanEncodingSmoothing(data, val_data, groupby, target, alpha=100, column_name=None):
    globalmean = data[target].mean() #global mean
    if column_name == None:
        column_name = _getMeanEncodingColumnName(groupby, target, 'mean')
    data[column_name] = data.groupby(groupby)[target].transform(lambda x: (x.mean() * x.count() + globalmean * alpha) / (x.count() + alpha))
    # Fill NaNs
    data[column_name].fillna(globalmean, inplace=True) #fill NaNs with global mean
    score = _getMeanEncodingCorrCoef(data, column_name, target)
    val_data = val_data.merge(data.groupby(groupby, as_index=False).agg({column_name : 'mean'}), on=groupby, how='left')
    return data, val_data, column_name, score

def MeanEncodingExpanding(data, val_data, groupby, target, column_name=None):
    globalmean = data[target].mean() #global mean
    if column_name == None:
        column_name = _getMeanEncodingColumnName(groupby, target, 'mean')
    cumsum = data.groupby(groupby)[target].cumsum() - data[target]
    cumcnt = data.groupby(groupby).cumcount()
    data[column_name] = cumsum/cumcnt
    # Fill NaNs
    data[column_name].fillna(globalmean, inplace=True) #fill NaNs with global mean
    score = _getMeanEncodingCorrCoef(data, column_name, target)
    val_data = val_data.merge(data.groupby(groupby, as_index=False).agg({column_name : 'mean'}), on=groupby, how='left')
    return data, val_data, column_name, score
```
